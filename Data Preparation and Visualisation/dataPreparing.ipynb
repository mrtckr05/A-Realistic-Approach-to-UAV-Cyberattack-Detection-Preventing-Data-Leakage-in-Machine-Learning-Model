**Data Handler Function**


*   Filters just wanted features from selected csvs.

*   Adds data file suffix to column names for prevent column name conflicts.

* Merges seperate csv files according to nearest timestamp with Pandas `merge_asof()`.

* Adds true class label with `data_class` argument.






```python
def handleData(data_class, data_list):
  data_class_choices = ['benign flight', 'gps jamming', 'gps spoofing']

  actuatorControl_features = ['timestamp', 'control[0]', 'control[1]', 'control[2]', 'control[3]', 'control[4]', 'control[5]', 'control[6]', 'control[7]']
  estimatorInnovationTestRatios_features = ['timestamp', 'gps_hvel[0]', 'gps_hvel[1]', 'gps_vvel', 'gps_hpos[0]', 'gps_hpos[1]', 'gps_vpos']
  estimatorInnovationVariances_features = ['timestamp', 'gps_hvel[0]', 'gps_hvel[1]', 'gps_vvel', 'gps_hpos[0]', 'gps_hpos[1]', 'gps_vpos']
  estimatorInnovations_features = ['timestamp', 'gps_hvel[0]', 'gps_hvel[1]', 'gps_vvel', 'gps_hpos[0]', 'gps_hpos[1]', 'gps_vpos']
  gpsPosition_features = ['timestamp', 'lat', 'lon', 'alt', 'alt_ellipsoid', 'eph', 'epv', 'hdop', 'vdop', 'noise_per_ms', 'jamming_indicator',
                             'vel_m_s', 'vel_n_m_s', 'vel_e_m_s', 'vel_d_m_s', 'cog_rad']

  localPosition_features = ['timestamp', 'x', 'y', 'z', 'vx', 'vy', 'vz', 'eph', 'epv', 'evh', 'heading']
  globalPosition_features = ['timestamp', 'lat', 'lon', 'alt', 'alt_ellipsoid', 'eph', 'epv']
  features_list = [actuatorControl_features, estimatorInnovationTestRatios_features, estimatorInnovationVariances_features,
                   estimatorInnovations_features, globalPosition_features, gpsPosition_features, localPosition_features]

  # Sadece istediğimiz özellikleri alıyoruz
  for i in range(len(features_list)):
    data_list[i] = data_list[i][features_list[i]]

  df_actuator = data_list[0][actuatorControl_features].copy()
  df_gps      = data_list[5][gpsPosition_features].copy()
  df_ratios   = data_list[1][estimatorInnovationTestRatios_features].copy()
  df_var      = data_list[2][estimatorInnovationVariances_features].copy()
  df_inn      = data_list[3][estimatorInnovations_features].copy()
  df_global   = data_list[4][globalPosition_features].copy()
  df_local    = data_list[6][localPosition_features].copy()

  # Sütun isimlerindeki çakışmaları önlemek için hangi csv'ye aitse onun suffix'ini sütun isminin sonuna ekliyoruz.
  # gps_position dataframesindeki isimler referans. Yani sonunda suffix eklemesi olmayanlar gps position verisi olarak anlaşılabilir.

  # Ratios
  df_ratios = df_ratios.rename(columns={c: c+'_ratio' for c in df_ratios.columns if c != 'timestamp'})

  # Variances
  df_var = df_var.rename(columns={c: c+'_var' for c in df_var.columns if c != 'timestamp'})

  # Innovations
  df_inn = df_inn.rename(columns={c: c+'_inn' for c in df_inn.columns if c != 'timestamp'})

  # Global
  df_global = df_global.rename(columns={c: c+'_glob' for c in df_global.columns if c != 'timestamp'})

  # Local
  df_local = df_local.rename(columns={c: c+'_loc' for c in df_local.columns if c != 'timestamp'})


  dfs_to_sort = [df_actuator, df_gps, df_ratios, df_var, df_inn, df_global, df_local]

  for df in dfs_to_sort:
      df.sort_values('timestamp', inplace=True)


  merged_df = pd.merge_asof(df_actuator, df_gps, on='timestamp', direction='nearest')
  merged_df = pd.merge_asof(merged_df, df_ratios, on='timestamp', direction='nearest')
  merged_df = pd.merge_asof(merged_df, df_var, on='timestamp', direction='nearest')
  merged_df = pd.merge_asof(merged_df, df_inn, on='timestamp', direction='nearest')
  merged_df = pd.merge_asof(merged_df, df_global, on='timestamp', direction='nearest')
  merged_df = pd.merge_asof(merged_df, df_local, on='timestamp', direction='nearest')


  if data_class in data_class_choices:
    merged_df['class'] = data_class

    print(f"Class label added successfully: {data_class}")
    print(merged_df[['timestamp', 'class']].sample(5))

  else:
    print(f"ERROR: '{data_class}' class name is wrong!")

  print(f"Total row number: {len(merged_df)}")
  print(f"Total column number: {len(merged_df.columns)}")

  print(merged_df.columns)
  return merged_df
```

**Creating New Data Frames with Sub Features**

```python
benign_data = handleData('benign flight', benign_data_list)
```

```python
jamming_data = handleData('gps jamming', jamming_data_list)
```

```python
spoofing_data = handleData('gps spoofing', spoofing_data_list)
```

# **Merging the DataFrames**

**Concatenation**

```python
master_data = pd.concat([benign_data, jamming_data, spoofing_data], ignore_index=True)
print(f"CONCATENATION Successfully Compleated! Total Row Number: {len(master_data)}")
```
